```python
# _ is sort of variable that is not used anywhere such as dummy variable
data['buying'],i = pd.factorize(data['buying'])
data['maint'],_ = pd.factorize(data['maint'])
data['doors'],_ = pd.factorize(data['doors'])
data['persons'],_ = pd.factorize(data['persons'])
data['lug_boot'],_ = pd.factorize(data['lug_boot'])
data['safety'],_ = pd.factorize(data['safety'])
```

```python
# factorize - label encoding
# get_dummies - one hot encoding
# in factorize - column['wishes'] have 3 different types of values ['hello','hi','gd_mrng'] is changed as [1,2,3]
# in get_dummies - column['Embarked'] have 3 different types of values['S','Q','C'] is changed as two columns as "Embarked_S","Embarked_Q"
# in above get_dummies for S - 'Embarked_S' column will be 1 if S exists else 0 ,for Q - 'Embarked_Q' column will be 1 if Q exist else 0 , for C both columns 'Embarked_Q','Embarked_S' both will be 0
# refer titanic dataset for better understanding
```

```python
x = data.iloc[:,0:-1]
y = data.iloc[:,1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
# train - 70% , test - 30% and so test_size = 0.3
# first 70% is for train and next 30% is for test
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree = tree.DecisionTreeClassifier(criterion='gini') # criterion = gini/entrophy
dtree.fit(x_train,y_train)
```

```python
y_train_pred = dtree.predict(x_train)
```

```python
accuracy = metrics.accuracy_score(y_train,y_train_pred)
print('Train accuracy : {:.2f}'.format(accuracy))
```

```python
y_test_pred = dtree.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test,y_test_pred)
print('Test accuracy : {:.2f}'.format(accuracy))
```
