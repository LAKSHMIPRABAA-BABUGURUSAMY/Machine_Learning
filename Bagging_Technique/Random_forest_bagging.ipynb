```python
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})
df["Married"]=df["Married"].map({"Yes":0,"No":1})
df["Education"]=df["Education"].map({"Graduate":0,"Not Graduate":1})
df["Dependents"].replace('3+',3,inplace=True)
df["Self_Employed"]=df["Self_Employed"].map({"Yes":1,"No":0})
df["Property_Area"]=df["Property_Area"].map({"Semiurban":1,"Urban":2,"Rural":3})
df["Loan_Status"]=df["Loan_Status"].map({"Y":1,"N":0})
```

```python
# label encoding used to assign values from 0 to n according to the number of unique values
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data1=lb.fit_transform(df['Loan_Status'])
data1
```

```python
# one hot encoding assign only 0 and 1 according to the presence of yes and no
from sklearn.preprocessing import OneHotEncoder
onehot_encoder=OneHotEncoder(sparse=False)
df1=np.array([df['Loan_Status']]).reshape(-1,1)
data=onehot_encoder.fit_transform(df1)
data
```

```python
# thus null value of each column is replaced by most repeated value(highest in number)(mode())
# thus null value of each column is replaced by mean value(mean())
rev_null=['Gender','Married','Dependents','Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History']
df[rev_null]=df[rev_null].replace({np.nan:df['Gender'].mode(),
                                  np.nan:df['Married'].mode(),
                                  np.nan:df['Dependents'].mode(),
                                  np.nan:df['Self_Employed'].mode(),
                                  np.nan:df['LoanAmount'].mode(),
                                  np.nan:df['Loan_Amount_Term'].mode(),
                                  np.nan:df['Credit_History'].mode()})
# the above code can alse obe replaced using fillna
# df["Gender"]=df["Gender"].fillna()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=['Gender','Married','Dependents','Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History','Loan_ID','Loan_Status']).values
y=df['Loan_Status'].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sta=StandardScaler()
x_train=sta.fit_transform(x_train)
x_test=sta.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc=RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(x_train,y_train)
```

```python
y_pred_train=rfc.predict(x_train)
print(classification_report(y_train,y_pred_train))
```

```python
y_pred_test=rfc.predict(x_test)
print(classification_report(y_test,y_pred_test))
```
